```python
#Importing necessary libraries
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
#added urls and headers 
url='https://www.imdb.com/chart/top/'
headers={"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/122.0.0.0 Safari/537.36 Edg/122.0.0.0"}
response=requests.get(url,headers=headers)

#Extracting the whole html
soup=BeautifulSoup(response.content,'html.parser')
soup2=BeautifulSoup(soup.prettify(), 'html.parser')
```

```python
#Extracting the details of the movie
movies = soup2.find('div', class_='ipc-metadata-list-summary-item__c')
movies.text.strip()
```

```python
#Extracting the year
year=soup2.find('span',class_='sc-b0691f29-8 ilsLEX cli-title-metadata-item')
year.text.strip()
```

```python
#we have to use the regular to get the ratings from the movie details
import re
rating=soup2.find('span',class_='sc-b0691f29-1 grHDBY')
rating = rating.text.strip()
rating_value = re.search(r'(\d+\.\d+)', rating)
if rating_value:
    print(rating_value.group())
else:
    print("Rating not found")
```

```python
#Creating the list to store the extracted values
movie_info=[]

#Getting the desired info by the running the loop and then appending it to movie_info
for movie in movies:
    title=movie.find('h3', class_='ipc-title__text').text.strip()
    
    rating_tag=movie.find('span',class_='sc-b0691f29-1 grHDBY')
    rating_text=rating_tag.text.strip() if rating_tag else "Rating not Found!"
    rating_value=re.search(r'(\d+\.\d+)', rating_text).group() if rating_text != "Rating not found" else "Rating not found"
    
    year_tag=movie.find('span',class_='sc-b0691f29-8 ilsLEX cli-title-metadata-item')
    year_text=year_tag.text.strip() if year_tag else 'Year not found!'
    
    #storing the movie details in key value format
    movie_info.append({'Title':title, 'Rating':rating_value, 'Year':year_text})
  
```

```python
#Printing the details
for movie in movie_info:
    print(movie)
```

```python
#TO save the data in the local machine
import csv

field_names=["Title","Rating","Year"]
path = r'C:\Users\devma\OneDrive\Desktop\Web Scraper\Web Scraper.csv'
with open(path, 'w', newline='', encoding='UTF8') as file:
   
    writer = csv.DictWriter(file, fieldnames=field_names)
    
    
    writer.writeheader()
    
    # Write the data rows
    for movie in movie_info:
        writer.writerow(movie)

print("CSV file created successfully!")
```
